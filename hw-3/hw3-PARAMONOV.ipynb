```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import warnings
import seaborn as sns
warnings.filterwarnings('ignore')
```

>## **0. Обработка датафрейма**

```python
df_hw = pd.read_csv('/Users/vsevolodparamonov/Downloads/Homework_2_data.csv') 
year = 2020
region = 'Челябинск'

if region == region:

    my_data = (df_hw[(df_hw.year == year) & (df_hw.region == region)]) 
    
my_data
```

> ## Задание 2 a

```python
import statsmodels.api as sm

my_data['high'] = my_data.educ.apply(lambda x: 1 if x == 2 else 0)

coefs_results = pd.DataFrame()

y = my_data.wage

X = np.hstack((np.ones((my_data.shape[0], 1)), my_data[['age', 'high', 'female', 'is_children', 'smoke']].values))

model = sm.OLS(y, X)

rez = model.fit()

coefs_results = pd.concat([coefs_results, rez.params, rez.bse, rez.tvalues, rez.pvalues,rez.conf_int(alpha=0.05)], axis=1)
coefs_results.columns = ['Est', 'SE','t_value', 'p_value','l_0.05', 'r_0.05']
coefs_results = coefs_results.reset_index().replace({'x1': 'age', 'x2':'high', 'x3':'female', 'x4':'is_children', 'x5':'smoke'}).rename(columns={'index':'coef'}).set_index('coef')
coefs_results['Significant'] = (coefs_results['l_0.05'] > 0) & (coefs_results['r_0.05'] > 0) | (coefs_results['l_0.05'] < 0) & (coefs_results['r_0.05'] < 0)
coefs_results
```

> ## Задание 2 б
>
> #### $$ \hat{wage_i} = \hat\beta_1 + \hat\beta_2 \times age_i + \hat\beta_3 \times high_i + \hat\beta_4 \times female_i + \hat\beta_5 \times is\_children_i + \hat\beta_6 \times smoke_i $$


```python
coefs_results = pd.DataFrame()

y = np.log(my_data.wage)

X = np.hstack((np.ones((my_data.shape[0], 1)), my_data[['age', 'high', 'female', 'is_children', 'smoke']].values))

model_ln_1 = sm.OLS(y, X)

rez_ln_1 = model_ln_1.fit()

coefs_results = pd.concat([coefs_results, rez_ln_1.params, rez_ln_1.bse, rez_ln_1.tvalues, rez_ln_1.pvalues, rez_ln_1.conf_int(alpha=0.05)], axis=1)
coefs_results.columns = ['Est', 'SE','t_value', 'p_value','l_0.05', 'r_0.05']
coefs_results = coefs_results.reset_index().replace({'x1': 'age', 'x2':'high', 'x3':'female', 'x4':'is_children', 'x5':'smoke'}).rename(columns={'index':'coef'}).set_index('coef')
coefs_results['Significant'] = (coefs_results['l_0.05'] > 0) & (coefs_results['r_0.05'] > 0) | (coefs_results['l_0.05'] < 0) & (coefs_results['r_0.05'] < 0)
coefs_results
```

> ## Задание 2 в
>
> #### $$ \widehat{ln(wage_i)} = \hat\beta_1 + \hat\beta_2 \times ln(age_i) + \hat\beta_3 \times high_i + \hat\beta_4 \times female_i + \hat\beta_5 \times is\_children_i + \hat\beta_6 \times smoke_i $$


```python
coefs_results = pd.DataFrame()

y = np.log(my_data.wage)

X = np.hstack((np.ones((my_data.shape[0], 1)), my_data[['age', 'high', 'female', 'is_children', 'smoke']].values))

X[:, 1] = np.log(X[:, 1])

model_ln_2 = sm.OLS(y, X)

rez_ln_2 = model_ln_2.fit()

coefs_results = pd.concat([coefs_results, rez_ln_2.params, rez_ln_2.bse, rez_ln_2.tvalues, rez_ln_2.pvalues, rez_ln_2.conf_int(alpha=0.05)], axis=1)
coefs_results.columns = ['Est', 'SE','t_value', 'p_value','l_0.05', 'r_0.05']
coefs_results = coefs_results.reset_index().replace({'x1': 'ln(age)', 'x2':'high', 'x3':'female', 'x4':'is_children', 'x5':'smoke'}).rename(columns={'index':'coef'}).set_index('coef')
coefs_results['Significant'] = (coefs_results['l_0.05'] > 0) & (coefs_results['r_0.05'] > 0) | (coefs_results['l_0.05'] < 0) & (coefs_results['r_0.05'] < 0)
coefs_results
```

> #### Сравнение моделей (1) и (2) с помощью $PE-теста$
>
> #### * (1) модель с wage, (2) - модель с ln(wage)

```python
import statsmodels.formula.api as smf

df_for_lin = pd.DataFrame()
df_for_log = pd.DataFrame()


data_with_log = my_data.copy()
data_with_log['log_wage'] = np.log(data_with_log['wage'])

model_lin = smf.ols('wage ~ age + high + female + is_children + smoke', data = data_with_log).fit()
model_log = smf.ols('log_wage ~ age + high + female + is_children + smoke', data = data_with_log).fit()

data_with_log['lin'] = np.log(model_lin.fittedvalues) - model_log.fittedvalues
data_with_log['log'] = model_lin.fittedvalues - np.exp(model_log.fittedvalues)
model_lin_add = smf.ols('wage ~ age + high + female + is_children + smoke + lin', data = data_with_log).fit()
model_log_add = smf.ols('log_wage ~ age + high + female + is_children + smoke+ log', data = data_with_log).fit()


df_for_lin = pd.concat([df_for_lin, model_lin_add.params, model_lin_add.bse, model_lin_add.tvalues, model_lin_add.pvalues, model_lin_add.conf_int(alpha=0.05)], axis=1)
df_for_lin.columns = ['Est', 'SE','t_value', 'p_value','l_0.05', 'r_0.05']
df_for_lin = df_for_lin.reset_index().replace({'x1': 'ln(age)', 'x2':'high', 'x3':'female', 'x4':'is_children', 'x5':'smoke'}).rename(columns={'index':'coef'}).set_index('coef')
df_for_lin['Significant'] = (df_for_lin['l_0.05'] > 0) & (df_for_lin['r_0.05'] > 0) | (df_for_lin['l_0.05'] < 0) & (df_for_lin['r_0.05'] < 0)

df_for_log = pd.concat([df_for_log, model_log_add.params, model_log_add.bse, model_log_add.tvalues, model_log_add.pvalues, model_log_add.conf_int(alpha=0.05)], axis=1)
df_for_log.columns = ['Est', 'SE','t_value', 'p_value','l_0.05', 'r_0.05']
df_for_log = df_for_log.reset_index().replace({'x1': 'ln(age)', 'x2':'high', 'x3':'female', 'x4':'is_children', 'x5':'smoke'}).rename(columns={'index':'coef'}).set_index('coef')
df_for_log['Significant'] = (df_for_log['l_0.05'] > 0) & (df_for_log['r_0.05'] > 0) | (df_for_log['l_0.05'] < 0) & (df_for_log['r_0.05'] < 0)
```
